```python
tagged_data = [TaggedDocument(words=word_tokenize(_d.lower()), tags=[str(i)]) for i, _d in enumerate(df_final['content'])]
```

```python
tagged_data[1]
```

# **Initializing doc2vec**

```python
model = gensim.models.doc2vec.Doc2Vec(vector_size=30, min_count=2, epochs=80)
```

# **Building the vocabulary of tagged data**

```python
model.build_vocab(tagged_data)
```

# **Training doc2vec**

```python
model.train(tagged_data, total_examples=model.corpus_count, epochs=80)
```

```python
model.save("/content/drive/MyDrive/IR/d2v.model")
```
